# Image Reconstruction by autoencoder

 Importing the necessary libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image #from image manipulation
```

defining the path of the dataset the dataset

```python
train_list = list(os.walk("./cats and dogs dataset/train"))[0][2]
```

```python
test_list = list(os.walk("./cats and dogs dataset/test"))[0][2]
```

creating an empty dataset of zeros  

```python
train_set = np.zeros((5000,128,128,3),dtype="float32")
```

```python
test_set = np.zeros((200,128,128,3),dtype="float32")
```

function to convert images into numpy arrays 

```python
def dataset_creation(path,image_list,image_size,empty_set,no_of_examples):
    for i in range(no_of_examples):
        img = Image.open(path+image_list[i])
        img = np.array(img.resize(image_size))/255 #since images were of different sizes it is better to make them of same size and normalize them 
        empty_set[i] = img
        print("Image no: "+str(i)+" done")
    return empty_set
```

```python
train_images_path = "./cats and dogs dataset/train/"
image_size = (128,128)
test_images_path = "./cats and dogs dataset/test/"
```

```python
train_set = dataset_creation(train_images_path,train_list,image_size,train_set,5000)
```

```python
test_set = dataset_creation(test_images_path,test_list,image_size,test_set,200)
```

```python
train_set.shape
```

```python
test_set.shape
```

defining an covolution autoencoder 

```python
def autoencoder(original_dim):
    input_layer = tf.keras.layers.Input(shape=original_dim)
    
    conv_encd1 = tf.keras.layers.Conv2D(16,(3,3),activation = "relu",padding = "same")(input_layer)
    pool_encd1 = tf.keras.layers.MaxPooling2D((2,2))(conv_encd1)
    conv_encd2 = tf.keras.layers.Conv2D(8,(3,3),activation = "relu",padding = "same")(pool_encd1)
    pool_encd2 = tf.keras.layers.MaxPooling2D((2,2))(conv_encd2)
    conv_encd3 = tf.keras.layers.Conv2D(8,(3,3),activation = "relu",padding = "same")(pool_encd2)
    bottle_neck = tf.keras.layers.MaxPooling2D((2,2))(conv_encd3)
    
    conv_decd1 = tf.keras.layers.Conv2D(8,(3,3),activation = "relu",padding = "same")(bottle_neck)
    upsmpl1 = tf.keras.layers.UpSampling2D((2,2))(conv_decd1)
    conv_decd2 = tf.keras.layers.Conv2D(8,(3,3),activation = "relu",padding = "same")(upsmpl1)
    upsmpl2 = tf.keras.layers.UpSampling2D((2,2))(conv_decd2)
    conv_decd3 = tf.keras.layers.Conv2D(16,(3,3),activation = "relu",padding = "same")(upsmpl2)
    upsmpl3 = tf.keras.layers.UpSampling2D((2,2))(conv_decd3)
    
    output_layer = tf.keras.layers.Conv2D(3,(3,3),activation = "sigmoid",padding="same")(upsmpl3)
    
    
    autoencoder = tf.keras.Model(input_layer,output_layer)
    return autoencoder
    
```

```python
autoencoder= autoencoder(original_dim=(128,128,3))
```

we are using adam optimizer with binary crossentropy as loss function 

```python
autoencoder.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
autoencoder.summary()
```

training the model 

```python
autoencoder.fit(train_set,train_set,epochs=10)
```

```python
plt.plot(autoencoder.history.history["loss"])
```

```python
result=autoencoder.predict(test_set[0].reshape(1,128,128,3))
```

original image 

```python
plt.imshow(test_set[0])
```

recontructed image 

```python
plt.imshow(result.reshape(128,128,3))
```

# Denoising autoencoder

adding noises to the images 

```python
noisy_train_set = train_set+(0.3*np.random.randn(len(train_set),128,128,3))
```

```python
noisy_test_set = test_set+(0.3*np.random.randn(len(test_set),128,128,3))
```

sample noisy image 

```python
plt.imshow(noisy_train_set[0])
```

creating the autoencoder 

```python
denoising_autoencoder = autoencoder(original_dim=(128,128,3))
```

```python
denoising_autoencoder.summary()
```

```python
denoising_autoencoder.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
denoising_autoencoder.fit(noisy_train_set,train_set,epochs=20)
```

```python
plt.plot(denoising_autoencoder.history.history["loss"])
```

```python
result = denoising_autoencoder.predict(noisy_test_set[0].reshape(1,128,128,3))
```

noisy image 

```python
plt.imshow(noisy_test_set[0])
```

reconstructed image  

```python
plt.imshow(result.reshape(128,128,3))
```
